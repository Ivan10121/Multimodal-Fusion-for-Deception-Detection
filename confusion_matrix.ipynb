```python
import numpy as np
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm,
                          labels=(0,1),
                          name_map={0:'Deceptive', 1:'Truthful'},
                          normalize=False,
                          title='Confusion Matrix',
                          cmap=plt.cm.Blues):

    # reconstruct class names according to the order of labels
    class_names = [ name_map[l] for l in labels ]

    cm = np.array(cm, dtype=np.int64)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)

    # Set axis ticks and labels in English
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels=class_names,
           yticklabels=class_names,
           ylabel='True label',
           xlabel='Predicted label',
           title=title)
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

    # Annotate cells
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i,
                    format(cm[i, j], fmt),
                    ha="center", va="center",
                    color='white' if cm[i, j] > thresh else 'black')

    fig.tight_layout()
    return fig, ax
```

```python
cm = [[19, 3],
      [8, 17]]

fig, ax = plot_confusion_matrix(cm, labels=(0,1))
```
